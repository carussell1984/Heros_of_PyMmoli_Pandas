## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create the bins for the age categories (I know the min and max from the previous excersize: 7 - 45)
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# Create the names for the four bins
age_brackets = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#add the categories to the table
purchase_data["Age Category"] = pd.cut(purchase_data["Age"], bins, labels=age_brackets)
purchase_data.head()
```

```python
#group the data by age category
group_by_age_cat_2 = purchase_data.groupby(["Age Category"])
group_by_age_cat_2
```

```python
#perform calc to get purchase count
purchase_count_by_age_2 = group_by_age_cat_2["Purchase ID"].count()
purchase_count_by_age_2
```

```python
#perform calculations to get the average purchase 
mean_purchase_price_by_age = group_by_age_cat_2["Price"].mean()
mean_purchase_price_by_age 
```

```python
#perform calculations to get the total purchased by age
total_purchase_by_age = group_by_age_cat_2["Price"].sum()
total_purchase_by_age 
```

```python
#print the above three calculations to a data frame, rename the columns

summary_age_purchase_draft = pd.DataFrame({"Purchase Count": purchase_count_by_age_2,
                                          "Average Purchase Price": mean_purchase_price_by_age,
                                          "Total Purchase Value": total_purchase_by_age})
summary_age_purchase_draft 
```

```python
#group by user sn and by price to get the average total spend by user
group_by_age_user_2 = purchase_data.groupby(["SN", "Age Category"])
group_by_age_user_2 
```

```python
#calculate the total purchases by user & then put thus in a data frame
sum_purchases_by_user_2 = pd.DataFrame(group_by_age_user_2["Price"].sum())
sum_purchases_by_user_2.head()
```

```python
#group by age category and then get the average total spend by user in the age category
group_by_age_user_3 = sum_purchases_by_user_2.groupby(["Age Category"])
ave_sum_purchase_by_user_age_cat = pd.DataFrame(group_by_age_user_3 ["Price"].mean())

#rename column
ave_sum_purchase_by_user_age_cat.rename(columns={"Price" : "Avg Total Purchase per Person"}, inplace=True)
ave_sum_purchase_by_user_age_cat
```

```python
#merge the two data frames above
summary_purchase_analysis_by_age = pd.concat([summary_age_purchase_draft, ave_sum_purchase_by_user_age_cat], axis=1, sort=False)

#format the three columns
summary_purchase_analysis_by_age["Average Purchase Price"] = summary_purchase_analysis_by_age["Average Purchase Price"].map("${:.2f}".format)
summary_purchase_analysis_by_age["Total Purchase Value"] = summary_purchase_analysis_by_age["Total Purchase Value"].map("${:.2f}".format)
summary_purchase_analysis_by_age["Avg Total Purchase per Person"] = summary_purchase_analysis_by_age["Avg Total Purchase per Person"].map("${:.2f}".format)
summary_purchase_analysis_by_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#step 1 is to group by user
purchase_data_1 = pd.read_csv(file_to_load)
top_spender_groupby = purchase_data_1.groupby(["SN"])
top_spender_groupby
```

```python
#calculation to find purchase count, set to frame, and rename column
top_spender_counts = top_spender_groupby["Purchase ID"].count().to_frame()
top_spender_counts.rename(columns={"Purchase ID" : "Total Counts"}, inplace=True)
top_spender_counts
```

```python
#calculation to find average purchase, set to frame, and rename column
top_spender_avg_purchase = top_spender_groupby["Price"].mean()
top_spender_avg_purchase = round(top_spender_avg_purchase, 2)

top_spender_avg_purchase = top_spender_avg_purchase.to_frame()
top_spender_avg_purchase.rename(columns={"Price" : "Average Purchase Price"}, inplace=True)
top_spender_avg_purchase
```

```python
#calculation to find total purchase, set to frame, and rename column
top_spender_total_purchase = top_spender_groupby["Price"].sum().to_frame()
top_spender_total_purchase.rename(columns={"Price" : "Total Purchase Value"}, inplace=True)
```

```python
#print the above on a data frame, sort by total purchase value, and then format the currency columns
summary_top_spenders = pd.concat([top_spender_counts, top_spender_avg_purchase, top_spender_total_purchase], axis = 1, sort=False)
summary_top_spenders 
```

```python
#sort by total purchase value
final_summary_top_spenders = summary_top_spenders.sort_values(["Total Purchase Value"], ascending=[False])
final_summary_top_spenders.head()
```

```python
#format currency columns
final_summary_top_spenders["Average Purchase Price"] = final_summary_top_spenders["Average Purchase Price"].map("${:.2f}".format)
final_summary_top_spenders["Total Purchase Value"] = final_summary_top_spenders["Total Purchase Value"].map("${:.2f}".format)
final_summary_top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#grab the columns needed
purchase_popular_item = purchase_data[["Item ID", "Item Name", "Price"]]
purchase_popular_item.head()
```

```python
#group by Item ID, and Item Name
popular_item_group_by = purchase_popular_item.groupby(["Item ID", "Item Name"])
```

```python
#basic calc for how many items were purchased, set to data frame, and rename column
popular_item_purchase_count = popular_item_group_by["Price"].count().to_frame()
popular_item_purchase_count.rename(columns={"Price" : "Purchase Count"}, inplace=True)

popular_item_purchase_count.head()
```

```python
#basic calc for item price, set to data frame, and rename column
popular_item_purchase_price = popular_item_group_by["Price"].mean().to_frame()
popular_item_purchase_price.rename(columns={"Price" : "Item Price"}, inplace=True)
popular_item_purchase_price.head()
```

```python
#basic calc for total purchase value by item, set to data frame, and rename column
popular_item_purchase_value = popular_item_group_by["Price"].sum().to_frame()
popular_item_purchase_value.rename(columns={"Price" : "Total Purchase Value"}, inplace=True)
popular_item_purchase_value.head()
```

```python
#merge above data frames and sort, and format price values
popular_items = pd.concat([popular_item_purchase_count, popular_item_purchase_price, popular_item_purchase_value], axis=1, sort=False)
popular_items_summary =popular_items.sort_values(["Purchase Count"], ascending=[False])
popular_items_summary["Item Price"] = popular_items_summary["Item Price"].map("${:.2f}".format)
popular_items_summary["Total Purchase Value"] = popular_items_summary["Total Purchase Value"].map("${:.2f}".format)
popular_items_summary.head()
```

```python
#Sort the above table by total purchase value, format currency columns
profitable_items_summary = popular_items.sort_values(["Total Purchase Value"], ascending=[False])
profitable_items_summary["Item Price"] = profitable_items_summary["Item Price"].map("${:.2f}".format)
profitable_items_summary["Total Purchase Value"] = profitable_items_summary["Total Purchase Value"].map("${:.2f}".format)
profitable_items_summary.head()
```
